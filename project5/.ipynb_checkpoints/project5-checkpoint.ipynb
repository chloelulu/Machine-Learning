```python
from sklearn import datasets,svm
from sklearn.model_selection import GridSearchCV
from sklearn import svm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations_with_replacement
from sklearn.model_selection import cross_val_score
import math
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

# 1. Binary classification

```python
data = pd.read_csv('ionosphere.data', sep=',',header=None, na_values=['?'])
#print(data.shape)
# print(data.isnull().sum())
data = data.dropna() # no NA in the df
print(data.shape)
data.describe()
```

```python
df = data.drop([1], axis=1) # since the 2nd column are all 0, drop this column
df.head()
```

# Separating features and labels

```python
print("Total number of labels: {}".format(df.shape[0]))
print("Number of male: {}".format(df[df[34] == 'g'].shape[0]))
print("Number of female: {}".format(df[df[34] == 'b'].shape[0]))
```

```python
X = df.drop(df.iloc[:,33:34], axis=1)
y = df.iloc[:,33:34] 
# Encode label category
gb_encoder = LabelEncoder()
y = gb_encoder.fit_transform(y)
```

# Data Standardisation

```python
# Scale the data to be between -1 and 1
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

# Splitting dataset into training set and testing set for better generalisation

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# Perform Grid search technique to find the best parameter

```python
# Coarse grid search
svm_model= SVC()
tuned_parameters = {
 'C': [2**-15, 2**-13, 2**-11, 2**-9, 2**-7, 2**-5, 2**-3, 2**-1, 2**1, 2**3, 2**5, 2**7, 2**9, 2**11, 2**13, 2**15] , 'gamma': [2**-15, 2**-13, 2**-11, 2**-9, 2**-7, 2**-5, 2**-3, 2**-1, 2**1, 2**3, 2**5, 2**7, 2**9, 2**11, 2**13, 2**15], 'kernel': ['rbf']}
model_svm = GridSearchCV(svm_model, tuned_parameters,cv=10,scoring='accuracy')
model_svm.fit(X_train, y_train)
print(model_svm.best_score_)
print(model_svm.best_params_)
```

```python
means = model_svm.cv_results_['mean_test_score']
stds = model_svm.cv_results_['std_test_score']
params = model_svm.cv_results_['params']
d = {}
idx = tuned_parameters['gamma']
for mean, std, params in zip(means, stds, params):
    if params['C'] not in d.keys():
        d[params['C']] = [mean]
    else:
        d[params['C']].append(mean)

df_score = pd.DataFrame(d, index=idx)
df_score
```

```python
# Use a finner resolution
svm_model= SVC()
tuned = {
    'C': (np.arange(2**-1, 2**3, 0.5)) , 'gamma': (np.arange(2**-7, 2**-3, 0.01)), 'kernel': ['rbf']
}
model_svm = GridSearchCV(svm_model, tuned,cv=10,scoring='accuracy')
model_svm.fit(X_train, y_train)
print("Best parameters set found:")
print(model_svm.best_params_, model_svm.best_score_)
```

```python
means = model_svm.cv_results_['mean_test_score']
stds = model_svm.cv_results_['std_test_score']
params = model_svm.cv_results_['params']
d = {}
idx = tuned['gamma']
for mean, std, params in zip(means, stds, params):
    if params['C'] not in d.keys():
        d[params['C']] = [mean]
    else:
        d[params['C']].append(mean)

df_score = pd.DataFrame(d, index=idx)
df_score
```

```python
y_true, y_pred = y_test, model_svm.predict(X_test)
print(classification_report(y_true, y_pred))
print("The accuracy score is %.4f" % accuracy_score(y_true, y_pred))
```

```python
data = df_score
yticks = data.index
keptticks = yticks[::int(len(yticks)/10)]
yticks = ['' for y in yticks]
yticks[::int(len(yticks)/10)] = keptticks
xticks = data.columns
keptticks = xticks[::int(len(xticks)/10)]
xticks = ['' for y in xticks]
xticks[::int(len(xticks)/10)] = keptticks
sns.heatmap(data,linewidth=0,yticklabels=yticks,xticklabels=xticks)
# This sets the yticks "upright" with 0, as opposed to sideways with 90.
plt.yticks(rotation=0) 
plt.show()
```

# 2. Multi-class problem

```python
# Use a finner resolution
svm_model= SVC()
tuned = {
    'C': (np.arange(2**3, 2**5, 0.5)) , 'gamma': (np.arange(2**-5, 2**-1, 0.05)), 'kernel': ['rbf']
}
model_svm = GridSearchCV(svm_model, tuned,cv=10,scoring='accuracy')
model_svm.fit(X_train, y_train)
print("Best parameters set found:")
print(model_svm.best_params_, model_svm.best_score_)

means = model_svm.cv_results_['mean_test_score']
stds = model_svm.cv_results_['std_test_score']
params = model_svm.cv_results_['params']

d = {}
idx = tuned['gamma']
for mean, std, params in zip(means, stds, params):
    if params['C'] not in d.keys():
        d[params['C']] = [mean]
    else:
        d[params['C']].append(mean)

df_score = pd.DataFrame(d, index=idx)
df_score
```

# 3. Image classification

```python
# Data import
Train = pd.read_csv('sat.trn',sep= '\s+', header = None)
Test = pd.read_csv('sat.tst',sep= '\s+', header = None)
print(sum(Train.isnull().sum()));print(sum(Test.isnull().sum()))
Train.head()
```

```python
Test.head()
Test.describe() # the 36th column is class
```

```python
X_Train = Train.drop(Train.iloc[:,36:37], axis=1)
X_Test = Test.drop(Test.iloc[:,36:37], axis=1)
y_train = Train.iloc[:,36:37]
y_test = Test.iloc[:,36:37]
# Scale the data to be between 0 and 1
scaler = MinMaxScaler()
scaler.fit(X_Train)
X_train = scaler.transform(X_Train)
scaler.fit(X_Test)
X_test = scaler.transform(X_Test)
```

```python
# Coarse grid search
svm_model= SVC()
tuned_parameters = {
 'C': [2**-5, 2**-3, 2**-1, 2**1, 2**3, 2**5, 2**7, 2**9, 2**11, 2**13, 2**15] , 'gamma': [2**-5, 2**-3, 2**-1, 2**1, 2**3, 2**5, 2**7, 2**9, 2**11, 2**13, 2**15], 'kernel': ['rbf']}
model_svm = GridSearchCV(svm_model, tuned_parameters,cv=5,scoring='accuracy')
model_svm.fit(X_train, y_train)
print("Best parameters set found:")
print(model_svm.best_params_, model_svm.best_score_)

means = model_svm.cv_results_['mean_test_score']
stds = model_svm.cv_results_['std_test_score']
params = model_svm.cv_results_['params']

d = {}
idx = tuned_parameters['gamma']
for mean, std, params in zip(means, stds, params):
    if params['C'] not in d.keys():
        d[params['C']] = [mean]
    else:
        d[params['C']].append(mean)

df_score = pd.DataFrame(d, index=idx)
df_score
```

```python
# Use a finner resolution
svm_model= SVC()
tuned = {
    'C': (np.arange(2**1, 2**5, 3)) , 'gamma': (np.arange(2**-3, 2**1, 0.375)), 'kernel': ['rbf']
}
model_svm = GridSearchCV(svm_model, tuned,cv=5,scoring='accuracy')
model_svm.fit(X_train, y_train)
print(model_svm.best_score_)
print(model_svm.best_params_)
means = model_svm.cv_results_['mean_test_score']
stds = model_svm.cv_results_['std_test_score']
params = model_svm.cv_results_['params']
d = {}
idx = tuned['gamma']
for mean, std, params in zip(means, stds, params):
    if params['C'] not in d.keys():
        d[params['C']] = [mean]
    else:
        d[params['C']].append(mean)

df_score = pd.DataFrame(d, index=idx)
df_score
```

```python
y_true, y_pred = y_test, model_svm.predict(X_test)
print(classification_report(y_true, y_pred))
print("The accuracy score is %.4f" % accuracy_score(y_true, y_pred))
```
