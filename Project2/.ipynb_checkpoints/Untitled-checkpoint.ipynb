```python
data = pd.read_csv('UTK-peers.csv', header=0, index_col=0, nrows=57)
#print(data.isnull().sum())
# drop three columns that are character, some variables contain 24,20 missing value were also deleted
data.drop(['HBC','2014 Med School','Vet School','Wall St. Jourl Rank','Med School Res $','AG Research ($000)'], axis=1, inplace=True)
clean_data = data[data.columns[2:]].replace('[\$,]', '', regex=True)
clean_data = clean_data[clean_data.columns[0:]].replace('-', np.nan, regex=True)
clean_data.iloc[:,1:] = clean_data.iloc[:,1:].astype(float)

# fill the NA value with mean value
clean_data.fillna(clean_data.mean(), inplace=True)
# normalize the data
clean_data = (clean_data- clean_data.mean()) / clean_data.std()
# count the number of NaN values in each column
print(clean_data.shape)
print(clean_data.isnull().sum())

clean_data_array = np.array(clean_data)
```

```python
# 6. Implement K-means clustering, report the number of iterations required for coveragence.
# Euclidian distance calculation
import math
def Euclidean_distance(u1, u2):
    ed = np.sum(np.square((u1-u2)))**0.5   
    return ed

def kmeans(k, data):
    # get initial k points
    np.random.seed(123)
    index = np.random.choice(data.shape[0],k)
    centroids = data[index, :]
    
    # assign each point to centriod
    assign_points = []
    for p in data:
        distance = [Euclidean_distance(p, c) for c in centroids]
        minimal_index = distance.index(min(distance))
        assign_points.append((minimal_index, p))

    ### converge standard
    sigma = 1/10**3
    diff = float('inf')
    
    ### iteration 
    iteration = 0
    while diff > sigma:
        
        old_centroids = centroids   # save old centroids
        
        # update the centriod 
        l = {} 
        for i, j in assign_points:
            if i not in l.keys():
                l[i] = [j]
            else:
                l[i] += [j]
        
        centroids_list = []        
        for key in sorted(l.keys()):
            centroids_list.append(sum(l[key])/len(l[key]))
        centroids = np.array(centroids_list)
        
        # iteration + 1
        iteration += 1
        
        # assign centroids
        assign_points = []
        for p in data:
            distance = [Euclidean_distance(p, c) for c in centroids]
            minimal_index = distance.index(min(distance))
            assign_points.append((minimal_index, p))
        #print(assign_points)
        # compute the difference between old entroids and new centroids
        diff = sum([Euclidean_distance(old, new) for old, new in zip(old_centroids, centroids)])
    print('Clustering compelte at %d' % iteration)
    return iteration, centroids, assign_points
```

```python
### function to compute dunn_index 
from collections import defaultdict
from itertools import combinations

def dunn_index(assign_points):
    
    # compute minimum inter distance
    min_inter_dist = []
    d = defaultdict(list)
    for i, p in assign_points:
        d[i].append(p)
    for k,j in combinations(list(d.keys()),2):
        min_dist = []
        c1, c2 = d[k], d[j]
        for p1 in c1:
            for p2 in c2:
                min_dist.append(Euclidean_distance(p1,p2))
        min_inter_dist.append(min(min_dist))

    # compute maximum intra dist, within a cluster
    max_intra_dist = []
    for k, v in d.items():
        max_dist = []
        if len(v)>=2:
            for p1, p2 in combinations(v,2):
                max_dist.append(Euclidean_distance(p1,p2))
        max_intra_dist.append(max(max_dist))
    
    #print(min(min_inter_dist), max(max_intra_dist))
    dunn_index = min(min_inter_dist) / max(max_intra_dist)
    return dunn_index
```

```python
## plot the dunn index for different cluster number
def plot_dunn(data, algorithm = 'kmeans', k=7):
    dunn_list = []
    for j in range(2,k):
        if algorithm == 'kmeans':
            iteration, centroids, assign_points = kmeans(j, data)
        else:
            assign_points = EM(data,k)
        dunn_list.append(dunn_index(assign_points))

    ## plot
    print(dunn_list)
    plt.plot(range(2,k), dunn_list, 'ro-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Dunn index')
    plt.show()
```

```python
## EM algorithm
from scipy.stats import multivariate_normal

def EM(X, k, eps=0.001, max_iters = 100):
       
    n, d = X.shape 
    # initialization
    mu = X[np.random.choice(n, k, False), :]
    Sigma= [np.eye(d)] * k
    w = [1./k] * k
    R = np.zeros((n, k))

    ### log_likelihoods
    log_likelihoods = []
    
    # Converage standard, <max_iters, or diff<eps        
    while len(log_likelihoods) < max_iters:

        # E - Step
        for i in range(k):
            R[:, i] = w[i] * multivariate_normal.pdf(X, mu[i], Sigma[i], allow_singular=True)

        log_likelihood = np.sum(np.log(np.sum(R, axis = 1)))

        log_likelihoods.append(log_likelihood)

        ## Normalize so that the responsibility matrix is row stochastic
        R = (R.T / np.sum(R, axis = 1)).T

        ## The number of datapoints belonging to each gaussian 
        h = np.sum(R, axis = 0)
        
        # get the label of each point       
        label_array = np.zeros_like(R)
        label_array[np.arange(len(R)), R.argmax(1)] = 1
        label = np.argmax(label_array, axis=1)
        
        # M Step
        ## calculate the new mean and covariance for each gaussian
        for i in range(k):

            ## means, MU
            mu[i] = 1. / h[i] * np.sum(R[:, i] * X.T, axis = 1).T
            x_mu = np.matrix(X - mu[i])

            ## covariances, Sigma
            Sigma[i] = np.array(1 / h[i] * np.dot(np.multiply(x_mu.T,  R[:, i]), x_mu))

            ## the probabilities, P
            w[i] = 1. / n * h[i]
        
        # check for onvergence
        if len(log_likelihoods) < 2 : continue
        if np.abs(log_likelihood - log_likelihoods[-2]) < eps: break
            
    return list(zip(label, X))
```

```python
def EM(X, k, eps=0.001, max_iters = 100):

    # n = number of data-points, d = dimension of data points        
    n, d = X.shape

    # randomly choose the starting centroids/means 
    ## as 3 of the points from datasets        
    mu = X[np.random.choice(n, k, False), :]

    # initialize the covariance matrices for each gaussians
    Sigma= [np.eye(d)] * k

    # initialize the probabilities/weights for each gaussians
    w = [1./k] * k

    # responsibility matrix is initialized to all zeros
    # we have responsibility for each of n points for eack of k gaussians
    R = np.zeros((n, k))

    ### log_likelihoods
    log_likelihoods = []

    
    # Iterate till max_iters iterations        
    while len(log_likelihoods) < max_iters:

        # E - Step

        ## Vectorized implementation of e-step equation to calculate the 
        ## membership for each of k -gaussians
        for i in range(k):
            R[:, i] = w[i] * multivariate_normal.pdf(X, mu[i], Sigma[i], allow_singular=True)

        ### Likelihood computation
        log_likelihood = np.sum(np.log(np.sum(R, axis = 1)))

        log_likelihoods.append(log_likelihood)

        ## Normalize so that the responsibility matrix is row stochastic
        R = (R.T / np.sum(R, axis = 1)).T

        ## The number of datapoints belonging to each gaussian 
        h = np.sum(R, axis = 0)
        
        # get the label of each point       
        N = np.zeros_like(R)
        N[np.arange(len(R)), R.argmax(1)] = 1
        label = np.argmax(N, axis=1)
        
        # M Step
        ## calculate the new mean and covariance for each gaussian by 
        ## utilizing the new responsibilities
        for i in range(k):

            ## means
            mu[i] = 1. / h[i] * np.sum(R[:, i] * X.T, axis = 1).T
            x_mu = np.matrix(X - mu[i])

            ## covariances
            Sigma[i] = np.array(1 / h[i] * np.dot(np.multiply(x_mu.T,  R[:, i]), x_mu))

            ## and finally the probabilities
            w[i] = 1. / n * h[i]
        
        # check for onvergence
        if len(log_likelihoods) < 2 : continue
        if np.abs(log_likelihood - log_likelihoods[-2]) < eps: break
    return list(zip(label, X))
```
